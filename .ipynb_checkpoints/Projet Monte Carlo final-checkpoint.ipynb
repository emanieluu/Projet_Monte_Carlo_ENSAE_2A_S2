## Données angulaires 

Une loi classique pour des données correspondant à des angles observés sur l’intervalle [−π, π] est la loi de von Mises, de densité: $ π(θ) = \frac{1}{Z(κ)} exp(κ cos(θ − µ))$ où κ > 0, µ ∈ [−π, π], et Z(κ) est une constante de normalisation qui n’admet pas d’expression explicite. Cette loi est souvent qualifiée de loi normale pour les angles, car $cos(θ) ≈ 1 − θ^2$

### 1. Algorithme acceptation-rejet

##### a) Avec la loi uniforme

```python
def gen_uni_pi(n=1):
    return np.random.uniform(-np.pi, np.pi, n)
```

```python
def unipi_density(x) :
    if x <= np.pi and x >= -np.pi:
        return 1/(2*np.pi)
    else:
        return 0
    
v_unipi_density = np.vectorize(unipi_density)

def von_mises_density(x, mu=0, kappa=1):
    return np.exp(kappa * np.cos(x - mu)) * (-np.pi <= x <= np.pi)

v_von_mises_density = np.vectorize(von_mises_density)
```

```python
y = np.linspace(-np.pi, np.pi)
M = np.max(v_von_mises_density(y, mu=0, kappa=1)/v_unipi_density(y))
```

```python
M
```

```python
def von_mises_ra(M=M, mu=0, kappa=1):
    while True :
        x = gen_uni_pi()
        u = rd.random()
        if u <= (1/M)*von_mises_density(x)/(unipi_density(x)):
            return x 
```

```python
sns.distplot([von_mises_ra() for i in range(20_000)])
plt.xlim(-np.pi, np.pi)
```

```python
#Plot the theoritical densities of f(x) et Mg(x)

x = np.linspace(-np.pi-0.01, np.pi+0.01, 300)
fig, ax = plt.subplots(1,1, figsize=(6,5))
ax.plot(x, [von_mises_density(i) for i in x], 'r-', label='theoretical VM')
ax.plot(x,[ M*unipi_density(i) for i in x], 'g-', lw=1, label='M x uniform')
fig.suptitle('Densités Théoriques')
plt.legend(prop={'size': 9})
plt.show()
fig.savefig('Densites_Théoriques')
```

```python
n = [10_000, 100_000, 1_000_000]
mu = 0
kappa = 1 

# Plot 3 figures with different number of iterations of the simulation (rescaled)

x = np.linspace(-np.pi, np.pi, 300)
fig, ax = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
i = 0

for nb in n:
    
    M = np.max(v_von_mises_density(x, mu=0, kappa=1)/v_unipi_density(x))
    
    res = [von_mises_ra(M,kappa,mu)[0] for i in range(nb)]
    
    ax[i].hist(res, bins=200, density=True)
    
    ax[i].plot(x, vonmises.pdf(x, kappa, mu), 'r-', lw=1, label='theoretical VM')

    ax[i].plot([-np.pi, np.pi], 2*[vonmises.pdf(x, kappa, mu).max()], 'g-', lw=1, label='uniform')

    ax[i].set_ylim([0., None])
    ax[i].set_xticks([-3.14, 0., 3.14])
    ax[i].title.set_text(f'n = {nb:.1e}')
    
    i += 1

ax[0].locator_params(axis="y", nbins=4)
ax[1].get_yaxis().set_visible(False)
ax[2].get_yaxis().set_visible(False)

fig.suptitle(f'von Mises  (\u03BC = {mu:.3f},  \u03BA = {kappa:.3f})')
plt.legend(prop={'size': 8})
plt.show()
```

###### Performance en fonction de κ et μ

```python
# Return acceptance rate 

def acceptance_rate_ra_uniform(kappa=1, mu=0,  n=100_000):
    data = gen_uni_pi(n=n)
    data = data - mu
    val = np.exp(kappa * (np.cos(data) - 1))
    von_mises = data[np.random.uniform(0, 1, n) <= val] 
    return 100 * von_mises.shape[0] / n
```

##### b) Avec la loi de Cauchy (bonus)

L'algorithme décrit dans le livre de Devroye présente certaines coquilles (le W est remplacé par Y ou le 1/cos est en vérité un arccos cf. flèches bleues), erreurs que nous avons corrigés avec d'autres articles traitant de ce sujet

```python
# Rejection function from Devroye book

from math import log1p

def von_mises_ra_cauchy(mu=0, kappa=1):
    r = 1 + np.sqrt(1 + 4 * kappa ** 2)
    rho = (r - np.sqrt(2 * r)) / (2 * kappa)
    s = (1 + rho**2) / (2 * rho)
    
    B = False
    while B == False :
        U = rd.random()
        V = 2*rd.random()-1

        Z = np.cos(np.pi * (2*U-1))
        W = (1 + s*Z)/(s+Z)
        Y = kappa*(s - W)

        x = Y*(2-Y) - V 
        if x >= 0 :
            B = True
            return np.sign(2*U-1)*np.arccos(W)
        if x<0 and log1p(np.abs(Y/V)) + 1 - Y >= 0:
            B = True 
            return np.sign(2*U-1)*np.arccos(W)
```

```python
sns.distplot([von_mises_ra_cauchy() for i in range(20_000)])
plt.xlim(-np.pi, np.pi)
```

```python
def wrapped_cauchy_density(x, kappa=1, mu=0):
    r = 1 + np.sqrt(1 + 4 * kappa ** 2)
    rho = (r - np.sqrt(2 * r)) / (2 * kappa)
    
    return (1/(2*np.pi))*(1-rho**2)/(1+ rho**2- 2*rho*np.cos(x))* (-np.pi- 0.01 <= x <= np.pi+ 0.01)
```

```python
x = np.linspace(-np.pi, np.pi, 300)
M2 = np.max([von_mises_density(i, mu=0, kappa=1)/wrapped_cauchy_density(i) for i in x])
```

```python
n = [10_000, 100_000, 1_000_000]
mu = 0
kappa = 1 

# Plot 3 figures with different number of iterations of the simulation 

x = np.linspace(-np.pi, np.pi, 300)
fig, ax = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
i = 0

for nb in n:
    
    res = [von_mises_ra_cauchy() for i in range(nb)]
    
    ax[i].hist(res, bins=200, density=True)
    
    ax[i].plot(x, vonmises.pdf(x, kappa, mu), 'r-', lw=1, label='theoretical VM')
    
    ax[i].plot(x, [1.2*wrapped_cauchy_density(i) for i in x], 'g-', lw=1, label="wrapped Cauchy" )

    ax[i].set_ylim([0., None])
    ax[i].set_xticks([-3.14, 0., 3.14])
    ax[i].title.set_text(f'n = {nb:.1e}')
    
    i += 1

ax[0].locator_params(axis="y", nbins=4)
ax[1].get_yaxis().set_visible(False)
ax[2].get_yaxis().set_visible(False)

fig.suptitle(f'von Mises  (\u03BC = {mu:.3f},  \u03BA = {kappa:.3f}) avec Cauchy')
plt.legend(prop={'size': 8})
plt.show()
fig.savefig('AR_Cauchy')
```

```python
def acceptance_rate_ra_cauchy(kappa=1, mu=0,  n=10_000):
    r = 1 + np.sqrt(1 + 4 * kappa ** 2)
    rho = (r - np.sqrt(2 * r)) / (2 * kappa)
    s = (1 + rho**2) / (2 * rho)
    
    c = 0
    for i in range(n):
        U = rd.random()
        V = 2*rd.random()-1

        Z = np.cos(np.pi * (2*U-1))
        W = (1 + s*Z)/(s+Z)
        Y = kappa*(s - W)

        x = Y*(2-Y) - V 
        if x >= 0 :
            c+=1
        elif x<0 and log1p(np.abs(Y/V)) + 1 - Y >= 0:
            c+=1
            
    return 100* c/n  
```

### 2. Algorithme de Metropolis

Proposer un algorithme de Metropolis simple pour simuler selon une telle loi; notamment proposer une règle simple pour que 
la performance de l’algorithme ne dépende pas des paramètres. 
(Expliquer ce que veut dire le mot “performance” pour un tel algorithme.)

Contrairement à l'algorithme de rejet, MCMC crée des va qui ne sont pas indépendantes les unes des autres

```python
# Gaussian

def metropolis(x, sig= 3, mu=0, kappa=1):
    y = x + sig * np.random.randn()
    r = von_mises_density(y, mu, kappa) / von_mises_density(x, mu, kappa)
    if np.random.rand() < r:
        return (y,1) 
    else:
        return (x,0)
```

```python
def simulation_metropolis(sig, kappa=1, mu=0,  N=100_000):
    x = np.zeros(N)
    for i in range(1,N):
        res = metropolis(x[i-1], sig, mu, kappa)
        x[i] = res[0]
    return x 
```

```python
def acceptance_rate_metropolis(sig, kappa=1, mu=0,  N=100_000):
    N_acceptance = 0
    x = np.zeros((N,2))
    for i in range(1,N):
        res = metropolis(x[i-1][0], sig, mu, kappa)
        x[i][0] = res[0]
        N_acceptance += res[1]
    return N_acceptance/N
```

```python
# Acceptance rate as a function of sig

res = [acceptance_rate_metropolis(sig, kappa=1, mu=0,  N=100_000) for sig in np.linspace(0.1, 10, 100)]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(np.linspace(0.1, 20, 100), res, 'b-')
ax.set_yticks([0, 0.2 ,0.4 ,0.6 ,0.8 ,1])
ax.grid(True, linewidth=0.5, color='grey', linestyle='-')
ax.set_xlabel("sig")
ax.set_ylabel("Acceptance rate")
ax.set_title("Acceptance rate as a function of sig")
plt.show()
```

```python
sig = [0.1, 1, 10]
mu = 0
kappa = 1 
N = 100_000

# Plot 3 figures with different number of iterations of the simulation 

xx = np.linspace(-np.pi, np.pi, 300)
fig, ax = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
i = 0

for s in sig:
    
    x = np.zeros(N)
    for j in range(1,N):
        res = metropolis(x[j-1], s, mu, kappa)
        x[j] = res[0]
    
    ax[i].hist(x, bins=200, density=True)
    
    ax[i].plot(xx, vonmises.pdf(xx, kappa, mu), 'r-', lw=1, label='theoretical VM')

    ax[i].set_ylim([0., None])
    ax[i].set_xticks([-3.14, 0., 3.14])
    ax[i].title.set_text(f'sig = {s}, acceptance_rate = {acceptance_rate_metropolis(s, kappa=1, mu=0, N=100_000)}')
    
    i += 1

ax[0].locator_params(axis="y", nbins=4)
ax[1].get_yaxis().set_visible(False)
ax[2].get_yaxis().set_visible(False)

fig.suptitle(f'von Mises  (\u03BC = {mu:.3f},  \u03BA = {kappa:.3f}) simulated with Metropolis')
plt.legend(prop={'size': 8})
plt.show()
```

##### Tentative d'implémentation de la règle simple 

```python
from scipy.misc import derivative

def inf_fisher_von_mises(k):

    def log_von_mises(x):
        return np.log(von_mises_density(x, kappa=k))
    
    def log_von_mises_diff_sq(x):
        return derivative(log_von_mises, x, dx=1e-6)**2
    
    r = vonmises.rvs(kappa, size=10000)
    
    res = [log_von_mises_diff_sq(x) for x in r]
    
    return sum(res)/len(res)

def simple_rule(kap):
    return (2.38)**2/inf_fisher_von_mises(kap)
```

### 3. MCMC-MLE

Utiliser l’approche MCMC-MLE pour estimer les paramètres $\mu$ et $\kappa$ et à
partir de données observées.

Méthode MCMC-MLE: méthode qui revient à remplacer dans une fonction
de log-vraisemblance une fonction des paramètres (ici Z) non calculable
par une approximation basée sur de l’importance sampling, où la loi de
proposition est la loi du modèle pour un paramètre fixe (bien choisi, à vous
de réfléchir comment). On maximise ensuite la log-vraisemblance ainsi
obtenue. Cette méthode est présentée notamment dans l’article de Geyer
& Thompson (1992).

###### On fixe $\phi = (\kappa_0, \mu_0) = (1,0)$ 
###### On considère $L(\theta_i, \kappa, \mu) = \prod_{i=1}^n \frac{\pi_{\kappa, \mu}(\theta_i)}{\pi_{1, 0}(\theta_i)}$, avec $\pi_{\kappa, \mu}(\theta) = \frac{1}{Z(\kappa)}\exp{\{\kappa \cos{(\theta_i - \mu)}\}}$

###### D'où $L(\theta_i, \kappa, \mu) = \prod_{i=1}^n \frac{\frac{1}{Z(\kappa)}\exp{\{\kappa \cos{(\theta_i - \mu)}\}}}{\frac{1}{Z(1)}\exp{\{\cos{(\theta_i)}\}}}$

###### En estimant $\kappa$ et $\mu$, on cherche à retrouver les paramètres de la loi inconnue $\pi_{\kappa, \mu}(\theta)$ qui aurait générée les données

###### On minimise ensuite $l_n(\theta_i, \kappa, \mu) = -\frac{1}{n} \times \log(L(\theta_i, \kappa, \mu))$

###### On a $l_n(\theta_i, \kappa, \mu) = -\frac{1}{n} \sum_{i=1}^n \{\kappa \cos(\theta_i - \mu) - \cos(\theta_i)\} + \log(\frac{Z(\kappa)}{Z(1)})$

###### D'après Geyer-Thompson, 1992, $ \frac{Z(\kappa)}{Z(1)} = \mathbb{E}_\phi [ \frac{\exp{\{\kappa \cos{(\theta - \mu)}\}}}{\exp{\{\cos{(\theta)}\}}}]$

###### Un estimateur de $\log(\frac{Z(\kappa)}{Z(1)})$ est donc $\log(\sum_{i=1}^n \frac{\exp{\{\kappa \cos{(\theta_i - \mu)}\}}}{\exp{\{\cos{(\theta_i)}\}}})$, où $\theta_i$ est distribué à partir de $\pi_{\phi}(\theta)$
###### Ce résultat repose sur l'ergodicité de la chaîne de Markov qui simule les données, c'est pourquoi on utilisera l'algorithme de Metropolis

#### Etape 1 : on code $l_n(\theta_i, \kappa, \mu)$ en remplaçant $\log(\frac{Z(\kappa)}{Z(1)})$ par son estimateur

```python
def log_likelihood(x, sample_inconnu, sample_fixe):
    mu = x[0]
    kappa = x[1]
    res_1 = kappa * np.cos(sample_inconnu - mu) - np.cos(sample_inconnu) # 
    res_2 = np.exp(kappa * np.cos(sample_fixe - mu) - np.cos(sample_fixe)) # terme de la somme de l'estimateur de log(Z(k)/Z(1))
    return -res_1.mean() + np.log(res_2.mean())

# Il y a 2 samples : sample_fixe généré par Metropolis avec kappa=1 et mu=0
#                    sample_inconnu généré par Metropolis avec kappa=0.8 et mu=0.9, les paramètres à trouver
```

#### Etape 2 : on minimise $l_n(\theta_i, \kappa, \mu)$ pour plusieurs échantillons et on considère la moyenne empirique pour approcher les vrais paramètres $\kappa$ et $\mu$

```python
params_estimated = np.zeros((500, 2))

for i in range(500):
    sample1 = simulation_metropolis(3, kappa=0.8, mu=0.9, N=1000) # paramètres à retrouver
    sample2 = simulation_metropolis(3, kappa=1, mu=0, N=1000) # paramètres fixés pour simplifier les calculs
    params_estimated[i] = minimize(log_likelihood, np.array([0., 1.]), args=(sample1, sample2)).x

mu_, kappa_ = params_estimated.mean(axis=0)
```
