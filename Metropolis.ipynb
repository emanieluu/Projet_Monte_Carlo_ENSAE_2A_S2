```python
import numpy as np
```

```python
def log_pi(x):
    return -0.5* x**2
```

```python
def metropolis(x, sig=0.1):
    y = x + sig * np.random.randn()
    log_r = log_pi(y) - log_pi(x)
    if np.log(np.random.rand()) < log_r:
        return y 
    else:
        return x
```

```python
N = 1000
x = np.empty(N)
x[0] = 1
for n in range(1,N):
    x[n] = metropolis(x[n-1], sig=2.)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x)
```
